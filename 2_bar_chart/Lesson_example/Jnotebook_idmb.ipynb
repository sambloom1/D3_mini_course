```python
import pandas as pd
```

```python
imdb_df = pd.read_csv("../imdb_top_1000.csv")
```

```python
imdb_clean = imdb_df[imdb_df["Genre"].str.contains("Mystery| Sci-Fi")]
imdb_clean = imdb_clean[imdb_clean["IMDB_Rating"] > 8.0]
```

```python
imdb_clean.tail()
```

```python
imdb_rename = imdb_clean.rename(columns={"Series_Title": "Movie"})
```

```python
imdb_final = imdb_rename[["Movie", "Genre", "Director", "IMDB_Rating", "Gross", "Meta_score"]]
```

```python
imdb_final.to_csv('../imdb_clean.csv', index=True)
```
